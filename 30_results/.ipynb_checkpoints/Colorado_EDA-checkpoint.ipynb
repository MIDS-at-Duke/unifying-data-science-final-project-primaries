```r
library(ggplot2)
```

```r
# Load data
colorado <- read.csv('../20_intermediate_files/turnout_colorado.csv', row.names=1)
head(colorado)
```

## EDA

### Age

```r
# 18-30 Year Olds
ggplot(colorado, aes(x=pop18.30, y=delta)) + 
    geom_point(mapping=aes(x=pop18.30, y=delta)) +
    geom_smooth(method = "lm") + 
    geom_hline(yintercept=0, color = 'green', alpha = .5) +
    ggtitle('Larger Younger Population Associated with Increased Turnout') +
    ylab('% point increase Voter Turnout') +
    xlab('% of eligible voters between 18-30') +
    theme_minimal()
```

```r
# 18-25 Year Olds
ggplot(colorado, aes(x=pop18.25, y=delta)) + 
    geom_point(mapping=aes(x=pop18.25, y=delta)) +
    geom_smooth(method = "lm") + 
    geom_hline(yintercept=0, color = 'green', alpha = .5) +
    ggtitle('Larger Younger Population Associated with Increased Turnout') +
    ylab('% point increase Voter Turnout') +
    xlab('% of eligible voters between 18-25') +
    
    theme_minimal()
```

```r
# Senior Citizens
ggplot(colorado, aes(x=pop65.85, y=delta)) + 
    geom_point(mapping=aes(x=pop65.85, y=delta)) +
    geom_smooth(method = "lm") + 
    geom_hline(yintercept=0, color = 'green', alpha = .5) +
    ggtitle('Opposite effect on senior citizens') +
    ylab('% point increase Voter Turnout') +
    xlab('% of eligible voters between 65-85') +
    theme_minimal()
```

### Race

```r
# White
ggplot(colorado, aes(x=White, y=delta)) + 
    geom_point(mapping=aes(x=White, y=delta)) +
    #geom_smooth(method = "lm") + 
    geom_hline(yintercept=0, color = 'green', alpha = .5) +
    ggtitle('Unclear Race Effect') +
    ylab('% point increase Voter Turnout') +
    xlab('% of eligible voters White') +
    theme_minimal()
```

### Population

```r
# Total Eligible voters
ggplot(colorado, aes(x=log(pop18.85), y=delta)) + 
    geom_point(mapping=aes(x=log(pop18.85), y=delta)) +
    geom_smooth(method = "lm") + 
    geom_hline(yintercept=0, color = 'green', alpha = .5) +
    ggtitle('Big Cities -> Increased Turnout') +
    ylab('% point increase Voter Turnout') +
    xlab('Log(% of eligible voters)') +
    theme_minimal()
```

## Quick Regression

```r
fit <- lm(delta ~ log(pop18.85) + pop18.25 + White + Black + Hispanic, data = colorado)
summary(fit)
```

```r
fit <- lm(delta ~ pop18.25 + White + Black + Hispanic, data = colorado)
summary(fit)
```
