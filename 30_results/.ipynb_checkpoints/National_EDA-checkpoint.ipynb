# National EDA
## Summary Stats for Final Report

```r
options(warn=-1) # Warnings disabled
library(dplyr)
library(ggplot2)
```

###  Load Data

```r
turnout <- read.csv('../20_intermediate_files/turnout_v3.csv', row.names=1)
head(turnout)
```

```r
# Set colors for viz
colors <- c('Control' = "blue", 'Treatment' = 'orange')
```

### Population

```r
# Kernal Density
ggplot() + 
    geom_density(data=turnout[turnout$treatment==FALSE,], 
                   mapping=aes(Pop, fill='Control'), alpha=.5) + 
    geom_density(data=turnout[turnout$treatment==TRUE,], 
                   mapping=aes(Pop, fill='Treatment'), alpha=.5) + 
    labs(x='Population', y='Density', title='Population Density by Treatment Group') +
    scale_fill_manual(values = c('Control' = "blue", 'Treatment' = 'orange')) + 
    theme_minimal()
```

```r
# Histogram
colors <- c('Control' = "blue", 'Treatment' = 'orange')
ggplot() + 
    geom_histogram(data=turnout[turnout$treatment==FALSE,], 
                   mapping=aes(Pop, fill='Control'), bins=15, alpha=.5) + 
    geom_histogram(data=turnout[turnout$treatment==TRUE,], 
                   mapping=aes(Pop, fill='Treatment'), bins=15, alpha=.5) + 
    labs(x='Population', y='Number of States', title='Population Density by Treatment Group') +
    scale_fill_manual(values = c('Control' = "blue", 'Treatment' = 'orange')) + 
    theme_minimal()
```
