```r
library(ggplot2)
library(dplyr)
```

```r
# Load data
colorado <- read.csv('../20_intermediate_files/turnout_colorado.csv', row.names=1)
head(colorado)
```

## EDA

```r
ggplot(colorado) +
    geom_histogram(mapping=aes(delta), bins =15, fill='lightblue') +
    xlab('% Point Change in Voter Turnout') +
    ylab('Number of Counties') +
    ggtitle('Change in Colorado County Voter Turnout 2016-2020') +
    theme_minimal() +
    theme(text = element_text(size=15))
```

### Age

```r
# 18-30 Year Olds
ggplot(colorado, aes(x=pop18.30, y=delta)) + 
    geom_point(mapping=aes(x=pop18.30, y=delta)) +
    geom_smooth(method = "lm") + 
    geom_hline(yintercept=0, color = 'green', alpha = .5) +
    ggtitle('Larger Younger Population Associated with Increased Turnout') +
    ylab('% point increase Voter Turnout') +
    xlab('% of eligible voters between 18-30') +
    theme_minimal() +
    theme(text = element_text(size=15))
```

```r
# 18-25 Year Olds
ggplot(colorado, aes(x=pop18.25, y=delta)) + 
    geom_point(mapping=aes(x=pop18.25, y=delta), size = 4, color = 'darkblue') +
    geom_smooth(method = "lm", alpha = .1) + 
    geom_hline(yintercept=0, color = 'black', alpha = .5, size = 2) +
    ggtitle('Colorado County Millennial Population and Voter Turnout') +
    ylab('% Point Chance in  Voter Turnout') +
    xlab('% of Population Between 18-25') +
    theme_minimal() +
    theme(text = element_text(size=15))
```

```r
# Senior Citizens
ggplot(colorado, aes(x=pop65.85, y=delta)) + 
    geom_point(mapping=aes(x=pop65.85, y=delta), size = 4, color = 'darkblue') +
    geom_smooth(method = "lm", alpha = .1) + 
    geom_hline(yintercept=0, color = 'black', alpha = .5, size = 2) +
    ggtitle('Colorado County Senior Population and Voter Turnout') +
    ylab('% Point Chance in  Voter Turnout') +
    xlab('% of Population 65+') +
    theme_minimal() +
    theme(text = element_text(size=15)) 
```

### Population

```r
# Total Eligible voters
ggplot(colorado, aes(x=log(pop18.85), y=delta)) + 
    geom_point(mapping=aes(x=log(pop18.85), y=delta)) +
    geom_smooth(method = "lm") + 
    geom_hline(yintercept=0, color = 'green', alpha = .5) +
    ggtitle('Big Cities -> Increased Turnout') +
    ylab('% point increase Voter Turnout') +
    xlab('Log(% of eligible voters)') +
    theme_minimal()
```

## Quick Regression

```r
fit <- lm(delta ~ pop18.25 + Black + Hispanic, data = colorado)
summary(fit)
```

```r
fit <- lm(delta ~ pop18.25 + White + Black + Hispanic, data = colorado)
summary(fit)
```
